```python
categorical_columns = ['Gender', 'PreferredCuisine', 'DiningOccasion', 'MealType', 'OnlineReservation', 'DeliveryOrder', 'LoyaltyProgramMember']
for column in categorical_columns:
    print(f"Unique values in {column}: {df[column].unique()}")
```

```python
#
sns.countplot(x=df['HighSatisfaction'] , palette='Set2', hue = 'HighSatisfaction',data = df)
plt.title('Distribution of HighSatisfaction')
plt.show()
```

```python
#meal timing(e.g., breakfast, lunch, dinner) has the highest average service rating

service_rating_by_meal_type = df.groupby('TimeOfVisit')['ServiceRating'].mean()

service_rating_by_meal_type.plot(kind='bar', color='lightblue')
plt.title('Average Service Rating by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Average Service Rating')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
#average spend vary by preferred cuisine

avg_spend_by_cuisine = df.groupby('PreferredCuisine')['AverageSpend'].mean().sort_values()

avg_spend_by_cuisine.plot(kind='bar', color='lightcoral')
plt.title('Average Spend by Preferred Cuisine')
plt.xlabel('Preferred Cuisine')
plt.ylabel('Average Spend (INR)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
#correlation between service rating and food rating

plt.scatter(df['ServiceRating'], df['FoodRating'], alpha=0.5, color='purple')
plt.title('Correlation between Service Rating and Food Rating')
plt.xlabel('Service Rating')
plt.ylabel('Food Rating')
plt.grid(True)
plt.tight_layout()
plt.show()

correlation = df['ServiceRating'].corr(df['FoodRating'])
print(f"Correlation between Service Rating and Food Rating: {correlation:.2f}")
```

```python
#
# Plot distributions for numerical features
numeric_columns = ['Age', 'Income', 'AverageSpend', 'WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']
for column in numeric_columns:
    sns.histplot(df[column].dropna(), kde=True , color='blue')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
#

for column in categorical_columns:
    sns.countplot(x=df[column] , hue=df[column], palette='Set2', data = df )
    plt.title(f'Count of {column}')
    plt.show()
```

```python
## Distribution of HighSatisfaction over different categorical features
for column in categorical_columns:
    sns.countplot(x=column, hue='HighSatisfaction', data=df)
    plt.title(f'Distribution of HighSatisfaction across {column}')
    plt.show()
```

```python
#
df_encoded = pd.get_dummies(df, drop_first=True)

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

```

```python
## Correlation with the target variable
corr_with_target = df_encoded.corr()['HighSatisfaction'].sort_values(ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x=corr_with_target.index, y=corr_with_target.values)
plt.xticks(rotation=90)
plt.title('Correlation of Features with HighSatisfaction')
plt.show()

```
