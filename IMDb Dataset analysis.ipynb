```python
#
sns.histplot(df['IMDb Rating'])
plt.title('Distribution of IMDb Ratings')
plt.show()
```

```python
#

sns.scatterplot(x='Duration (minutes)', y='IMDb Rating', data=df)
plt.title('Scatterplot of Duration vs. IMDb Rating')
plt.show()
```

```python
#

sns.lineplot(x='Year', y='IMDb Rating', data=df)
plt.title('Lineplot of Year vs. IMDb Rating')
plt.show()
```

```python
#

sns.relplot(x='IMDb Rating', y='MetaScore', hue='Genre', data=df)
plt.title('Relplot of IMDb Rating and MetaScore by Genre')
plt.show()
```

```python
#

sns.barplot(x='Genre', y='IMDb Rating', data=df)
plt.title('Barplot of Average IMDb Rating by Genre')
plt.show()
```

```python
#

sns.boxplot(x='Genre', y='IMDb Rating', data=df)
plt.title('Boxplot of IMDb Rating by Genre')
plt.show()
```

```python
#

sns.countplot(x='Genre', data=df)
plt.title('Countplot of Movies by Genre')
plt.show()
```

```python
#

sns.catplot(x='Certificates', y='IMDb Rating', kind='violin', data=df)
plt.title('Catplot of IMDb Rating by Certificates')
plt.show()
```

```python
#

sns.violinplot(x='Genre', y='IMDb Rating', data=df)
plt.title('Violinplot of IMDb Rating by Genre')
plt.show()
```

```python
#

sns.displot(df['IMDb Rating'], kde=True)
plt.title('Distribution Plot of IMDb Ratings')
plt.show()
```

```python
#

sns.jointplot(x='IMDb Rating', y='MetaScore', data=df, kind='scatter')
plt.show()
```

```python
#

sns.lmplot(x='MetaScore', y='IMDb Rating', data=df)
plt.title('Linear Model Plot - IMDb Rating vs MetaScore')
plt.show()
```

```python
#

sns.pairplot(df)
plt.title('Pairplot of Movie Attributes')
plt.show()
```
