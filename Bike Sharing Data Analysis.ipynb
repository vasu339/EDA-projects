```python
day_categorical_columns = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
day_numerical_columns = ['temp', 'atemp', 'hum', 'windspeed','casual','registered','cnt']
```

```python
hour_categorical_columns = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']
hour_numerical_columns = ['hr','temp', 'atemp', 'hum', 'windspeed','casual','registered','cnt']
```

```python
for column in hour_numerical_columns:
    sns.histplot(hour_df[column].dropna(), kde=True , color='blue')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
for column in day_numerical_columns:
    sns.histplot(day_df[column].dropna(), kde=True , color='green')
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
for column in day_categorical_columns:
    sns.countplot(x=day_df[column] , hue=day_df[column], palette='Set2', data = day_df )
    plt.title(f'Count of {column}')
    plt.show()
```

```python
for column in hour_categorical_columns:
    sns.countplot(x=hour_df[column] , hue=hour_df[column], palette='Set2', data = hour_df )
    plt.title(f'Count of {column}')
    plt.show()
```

```python
sns.pairplot(day_df[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']])
```

```python
#

columns = [ 'season', 'yr', 'mnth', 'hr', 'holiday', 'weekday',
       'workingday', 'weathersit',  'casual', 'registered', 'cnt']
corr_matrix = hour_df[columns].corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',fmt = '0.2f')
plt.show()
```

```python
## plotting histogram of correlation with target column i.e. cnt
sns.histplot(corr_matrix,
            x='cnt',
            bins=10,
            color='gray')
```

```python
sns.lineplot(hour_df.groupby(by= ['hr'])['cnt'].sum(),color = 'orange',linewidth = 3)
plt.fill_between(x =hour_df['hr'].unique(),y1 =hour_df.groupby(by=['hr'])['cnt'].sum(), color='blue', alpha=0.2)
plt.show()
# people are mostly renting bicycle during office opening time mostly at 8-9 AM and then at office closing time around 4-5 PM
# Some people are renting bicycles even at MidNight
```

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'], errors='coerce')
day_df['weekday'] = day_df['dteday'].dt.day_name()  # Extract day of the week from date
day_counts = day_df['weekday'].value_counts()
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Distribution of Rentals by Day of the Week (Daily Data)')
plt.show()
```

```python
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'], errors='coerce')
hour_df['weekday'] = hour_df['dteday'].dt.day_name() 
hour_counts = hour_df['weekday'].value_counts()
plt.pie(hour_counts, labels=hour_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Distribution of Rentals by Hour of the Day (Hourly Data)')
plt.show()
```
