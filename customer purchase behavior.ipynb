```python
#Time Spent on Website by Product Category

# Plotting
sns.lineplot(x='ProductCategory', y='TimeSpentOnWebsite', data=df, marker='o')

plt.title('Time Spent on Website by Product Category')
plt.xlabel('ProductCategory')
plt.ylabel('TimeSpentOnWebsite (minutes)')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['Electronics', 'Clothing', 'Home Goods', 'Beauty', 'Sports'])

plt.show()
```

```python
# Histogram: Annual Income
plt.figure(figsize=(8, 6))
sns.histplot(df['AnnualIncome'], kde=True, bins=10, color='blue')
plt.title('Distribution of Annual Income')
plt.xlabel('AnnualIncome')
plt.ylabel('Frequency')
plt.show()
```

```python

# Grouping the data by Loyalty Program and calculating the mean time spent
grouped_df = df.groupby('LoyaltyProgram')['TimeSpentOnWebsite'].mean().reset_index()

# Plotting
sns.barplot(x='LoyaltyProgram', y='TimeSpentOnWebsite', data=grouped_df, palette='Blues', hue = 'LoyaltyProgram')

plt.title('Average Time Spent on Website by Loyalty Program')
plt.xlabel('LoyaltyProgram')
plt.ylabel('Average TimeSpentOnWebsite (minutes)')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])

plt.show()
```

```python
#

# Creating age groups
df['Age Group'] = pd.cut(df['Age'], bins=[20, 35, 70], labels=['20 to 35', '35 to 70'], right=False)

# Counting the number of customers by Gender and Age Group with observed=True to avoid the warning
count_df = df.groupby(['Gender', 'Age Group'], observed=True).size().unstack(fill_value=0).reset_index()

# Melting the DataFrame for better visualization
melted_df = pd.melt(count_df, id_vars='Gender', value_vars=['20 to 35', '35 to 70'], var_name='Age Group', value_name='Count')

# Plotting
sns.barplot(x='Age Group', y='Count', hue='Gender', data=melted_df, palette='Set2')

plt.title('Customer Count by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.xticks(ticks=[0, 1], labels=['20 to 35', '35 to 70'])
plt.legend(title='Gender', labels=['Male', 'Female'])

plt.show()
```

```python
#

df.groupby('ProductCategory').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#

df.value_counts('LoyaltyProgram')
sns.barplot(x='Gender',y='LoyaltyProgram',data=df)
```

```python
#
LoyaltyProgram_counts = df['LoyaltyProgram'].value_counts()
# Plot the pie chart
plt.pie(LoyaltyProgram_counts, labels=LoyaltyProgram_counts.index, autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=140)
plt.title('Distribution of LoyaltyProgram')
plt.show()
```

```python
#

plt.figure(figsize=(8, 4))
sns.boxplot(x='ProductCategory', y='NumberOfPurchases', data=df)
plt.title('NumberOfPurchases by Product Category')
plt.show()
```

```python
#

plt.figure(figsize=(8, 4))
sns.histplot(df['NumberOfPurchases'], bins=30, kde=True)
plt.title('Distribution of NumberOfPurchases')
plt.show()
```

```python
#

# Creating the relplot
sns.relplot(
    x='AnnualIncome', 
    y='TimeSpentOnWebsite',
    hue='PurchaseStatus', 
    style='Gender',
    size='LoyaltyProgram',
    sizes=(50, 200),
    col='ProductCategory',
    data=df
)

plt.suptitle('Relationship Between Annual Income and Time Spent On Website by Purchase Status and Gender', y=1.02)
plt.show()
```

```python
# Relplot: Time Spent on Website vs Number of Purchases by Product Category
sns.relplot(x='TimeSpentOnWebsite', y='NumberOfPurchases', hue='ProductCategory', style='LoyaltyProgram', data=df)
plt.title('Time Spent on Website vs Number of Purchases')
plt.xlabel('TimeSpentOnWebsite')
plt.ylabel('NumberOfPurchases')
plt.show()
```
