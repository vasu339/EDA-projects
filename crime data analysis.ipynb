```python
data = df.drop_duplicates()
data
```

```python
#

d=data["Vict Age"]<18
data.where(d).dropna()
```

```python


# Filter the data
under_18_data = data[data["Vict Age"] < 18]

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(under_18_data["Vict Age"], bins=10, kde=False, color='skyblue')
plt.title('Distribution of Victim Ages Under 18')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=under_18_data["Vict Age"], color='lightgreen')
plt.title('Boxplot of Victim Ages Under 18')
plt.xlabel('Age')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#

# Count of each crime type
top_n = 15
top_crimes = data['Crm Cd Desc'].value_counts().nlargest(top_n)
sns.barplot(x=top_crimes.values, y=top_crimes.index)
plt.title(f'Top {top_n} Most Common Crime Types', fontsize=16)
plt.xlabel('Number of Incidents', fontsize=14)
plt.ylabel('Crime Type', fontsize=14)
plt.show()
```

```python
#

sns.countplot(y='AREA NAME', data=data, order=data['AREA NAME'].value_counts().index)
plt.title('Crimes by Area')
plt.show()
```

```python
#
data['AREA NAME'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
```

```python
#

#Distribution of Crime Status solved vs unsolved
sns.countplot(data=data, x='Status Desc', color= 'skyblue')
plt.title('Distribution of Crime Status')
plt.xlabel('Status Description')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
#plt.figure(figsize=(12, 6))
sns.histplot(data['Vict Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#

# Top Crime Types
sns.countplot(data=data, y='Crm Cd Desc',color='red', order=data['Crm Cd Desc'].value_counts().index[:5])
plt.title('Distribution of Top 5 Crime Types')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Description')
plt.show()
```

```python

# draw lineplot 
sns.lineplot(x="Vict Age", y="Weapon Desc", data=data)
```

```python
## Select numeric columns for correlation analysis
numeric_data = data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Choose two numeric columns for scatterplot
# Example: Scatterplot of 'Vict Age' vs crm cd
sns.scatterplot(x='Vict Age', y='Crm Cd', data=data)
plt.title('Scatterplot of Vict Age vs. Crm Cd')
plt.xlabel('Vict Age')
plt.ylabel('Crm Cd')
plt.show()
```

```python
# Select numeric columns for pairplot
numeric_columns = ['Vict Age']  # Add more numeric columns if available
numeric_data = data[numeric_columns]

# Create pairplot
sns.pairplot(numeric_data)
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```
